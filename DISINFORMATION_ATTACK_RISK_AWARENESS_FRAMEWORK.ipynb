```python
import csv
import pandas as pd

file = open('DISARM_countermeasures.csv')
type(file)
csvreader = csv.reader(file)
header = []
header = next(csvreader)

rows = []
for row in csvreader:
  rows.append(row)

def get_field_name(row, fieldname, sep):
  if fieldname == "tactic":
    y = row['tactic'].split(sep)
    x = y
    y.pop(0)
    y = ' '.join(y)
    if len(x) > 1:
      x[1] = y
  else:
    x=row[fieldname].split(sep)
  if len(x) > 1:
    return x[1]
  return ""

def get_field_id(row, fieldname, sep):
  x=row[fieldname].split(sep)
  return x[0]

countermeasures = pd.DataFrame(rows, columns=header)
countermeasures = countermeasures[["disarm_id", "name", "metatechnique", "summary", "actortypes", "tactic", "tags", "techniques"]]
countermeasures['tacticname'] = countermeasures.apply (lambda row: get_field_name(row, "tactic", " "), axis=1)
countermeasures['tactic'] = countermeasures.apply (lambda row: get_field_id(row, "tactic", " "), axis=1)
countermeasures['actortypesname'] = countermeasures.apply (lambda row: get_field_name(row, "actortypes", "-"), axis=1)
countermeasures['actortypes'] = countermeasures.apply (lambda row: get_field_id(row, "actortypes", "-"), axis=1)
countermeasures['techniquename'] = countermeasures.apply (lambda row: get_field_name(row, "techniques", "-"), axis=1)
countermeasures['techniques'] = countermeasures.apply (lambda row: get_field_id(row, "techniques", "-"), axis=1)
countermeasures = countermeasures.reset_index()
countermeasures.head(10)
```

```python
from owlready2 import *

#onto = get_ontology("http://test.org/disarm_saw.owl")
onto = get_ontology("STO.owl").load()

basesto = onto.get_namespace("http://vistology.com/ont/2008/STO/STO.owl#")
ind_class = basesto.Individual
```

classes

```python
with onto:
     class Phases(ind_class):
         pass
     class Tactics(Phases):
         pass
     class Techniques(Tactics):
         pass
     class Countermeasures(Tactics):
         pass
     class Tools(ind_class):
         pass
     class Incidents(ind_class):
         pass
     class Incidents_techniques(ind_class):
         pass
     class is_in_phase(ObjectProperty):
         domain    = [Tactics]
         range     = [Phases]
     class is_in_tactic(ObjectProperty):
         domain    = [Techniques]
         range     = [Tactics]
     class is_used_in(ObjectProperty):
         domain    = [Incidents_techniques]
         range     = [Incidents]
     class uses_technique(ObjectProperty):
         domain    = [Countermeasures, Incidents_techniques]
         range     = [Techniques]
     class uses_tactic(ObjectProperty):
         domain    = [Countermeasures]
         range     = [Tactics]
     class has_id(DataProperty, FunctionalProperty):
         range = [str]
     class has_name(DataProperty, FunctionalProperty):
         range = [str]
     class has_summary(DataProperty, FunctionalProperty):
         range = [str]
     class has_externalgroup(DataProperty, FunctionalProperty):
         range = [str]
     class has_url(DataProperty, FunctionalProperty):
         range = [str]
     class has_category(DataProperty, FunctionalProperty):
         range = [str]
     class has_disinformation_use(DataProperty, FunctionalProperty):
         range = [str]
     class has_function(DataProperty, FunctionalProperty):
         range = [str]
     class has_artifacts(DataProperty, FunctionalProperty):
         range = [str]
     class has_found_in_country(DataProperty, FunctionalProperty):
          range = [str]
     class has_attributions_seen(DataProperty, FunctionalProperty):
          range = [str]
```

```python
#Add Relation subclass
rel_class = basesto.Relation
att_class = basesto.Attribute
time_class = basesto.Time
loc_class = basesto.Location
with onto:
  class hasUrl(rel_class):
    pass
  class happenedIn(rel_class):
    pass
  class isAttributedTo(rel_class):
    pass
  class happenedInContext(rel_class):
    pass
  class hasPhase(rel_class):
    pass
  class hasTactic(rel_class):
    pass
  class hasTechnic(rel_class):
    pass
  class hasCountermeasure(rel_class):
    pass
  class attName(att_class):
    pass
  class attSummary(att_class):
    pass
  class yearStarted(time_class):
    pass
  class whenAdded(time_class):
    pass
  class country(loc_class):
    pass
```

```python
#Incident Holds basic information about an incident (name, description, start date, objective etc.)
#Observable A url or file that has been observed in an incident.
#Channels: Any online or offline communication channel (a website, social media profile or page, TV station etc.). (extension) Channels publish observables.
#Identity Individuals, organisations or locations including countries. Usually encoding the targets of incidents.
#Threat Actor Holds information about a threat actor
#Event (extension) Describes a real life event like an election, show, anniversary etc. to provide context in which incidents can take place.
#Vulnerability Describes a vulnerability that was exploited to make an incident work.
#Language (extension) Which language(s) was (were) used in an incident
#Attack Pattern Describes manipulative techniques (TTP) used to conduct an attack
#Course of Action Describes countermeasures to incidents.
#Narrative (extension) Describes narratives used in incidents. Narratives can be nested and be represented as meta- and sub-narratives
for index, row in incidents.iterrows():
  sitname = basesto.Situation(row['name'])

  #Incident
  #incident = basesto.ElementaryInfon("incident_"+str(index))
  #sitname.supportedInfon.append(incident)

  #SITUATION ATTRIBUTES
  #name
  att = basesto.attName("att_name_"+str(index))
  sitname.hasAttribute.append(att)
  val = basesto.Value("val_name_"+str(index))
  att.hasAttributeValue.append(val)
  val.attributeValue.append(row['name'])

  #Description
  att = basesto.attSummary("att_desc_"+str(index))
  sitname.hasAttribute.append(att)
  val = basesto.Value("val_desc_"+str(index))
  att.hasAttributeValue.append(val)
  val.attributeValue.append(row['summary'])

  #start date
  att = basesto.yearStarted("att_start_date_"+str(index))
  sitname.hasAttribute.append(att)
  val = basesto.Value("val_start_date_"+str(index))
  att.hasAttributeValue.append(val)
  val.attributeValue.append(row['year_started'])

  #when added
  att = basesto.whenAdded("att_when_added_"+str(index))
  sitname.hasAttribute.append(att)
  val = basesto.Value("val_when_added_"+str(index))
  att.hasAttributeValue.append(val)
  val.attributeValue.append(row['when_added'])

  #found in country
  att = basesto.country("att_found_in_country_"+str(index))
  sitname.hasAttribute.append(att)
  val = basesto.Value("val_found_in_country_"+str(index))
  att.hasAttributeValue.append(val)
  val.attributeValue.append(row['found_in_country'])

  #SITUATION INFONS
  #Observable

  r = list(default_world.sparql("""
           SELECT ?inc ?url
           { ?inc STO:has_name \'""" + row['name'] + """\' .
             ?inc STO:has_url ?url .
           }
    """))
  if len(r) != 0:
    for index2, item in enumerate(r):
      #Add observables
      ob_ind = basesto.Individual("observable_ind"+str(index)+"_"+str(index2))
      att = basesto.Attribute("att_observable_"+str(index))
      ob_ind.hasAttribute.append(att)
      val = basesto.Value("val_ob_"+str(index))
      att.hasAttributeValue.append(val)
      val.attributeValue.append(item[1])

      observable_inf = basesto.ElementaryInfon("url_"+str(index)+"_"+str(index2))
      sitname.supportedInfon.append(observable_inf)
      observable_rel = basesto.hasUrl("has_url_"+str(index)+"_"+str(index2))
      observable_inf.relation = observable_rel
      observable_inf.anchor1 = item[0]
      observable_inf.anchor2 = ob_ind

  #Identity individuals
  r = list(default_world.sparql("""
           SELECT ?inc ?location
           { ?inc STO:has_name \'""" + row['name'] + """\' .
             ?inc STO:has_found_in_country ?location .
           }
    """))
  if len(r) != 0:
    for index2, item in enumerate(r):
      #Add individuals
      loc_ind = basesto.Individual("location_individual_"+str(index)+"_"+str(index2))
      att = basesto.Location("att_loc_"+str(index))
      loc_ind.hasAttribute.append(att)
      val = basesto.Value("val_loc_"+str(index))
      att.hasAttributeValue.append(val)
      val.attributeValue.append(item[1])

      #create infon to relate the infon to the individual
      loc_inf = basesto.ElementaryInfon("location_identity_"+str(index)+"_"+str(index2))
      sitname.supportedInfon.append(loc_inf)
      loc_rel = basesto.happenedIn("happened_in_"+str(index)+"_"+str(index2))
      loc_inf.relation = loc_rel
      loc_inf.anchor1 = item[0]
      loc_inf.anchor2 = loc_ind


  #Threat actor
  r = list(default_world.sparql("""
           SELECT ?inc ?att_seen
           { ?inc STO:has_name \'""" + row['name'] + """\' .
             ?inc STO:has_attributions_seen ?att_seen .
           }
    """))
  if len(r) != 0:
    for index2, item in enumerate(r):
      #Add attibute
      attribution_ind = basesto.Individual("threat_actor_individual_"+str(index)+"_"+str(index2))
      att = basesto.Attribute("att_attribution_"+str(index))
      attribution_ind.hasAttribute.append(att)
      val = basesto.Value("val_attribution_"+str(index))
      att.hasAttributeValue.append(val)
      val.attributeValue.append(item[1])

      #Add attribution to individuals
      att_seen_inf = basesto.ElementaryInfon("attribution_seen_"+str(index)+"_"+str(index2))
      sitname.supportedInfon.append(att_seen_inf)
      att_seen_rel = basesto.isAttributedTo("is_attributed_to_"+str(index)+"_"+str(index2))
      att_seen_inf.relation = att_seen_rel
      att_seen_inf.anchor1 = item[0]
      att_seen_inf.anchor2 = attribution_ind

  #Event
  r = list(default_world.sparql("""
           SELECT ?inc ?summary
           { ?inc STO:has_name \'""" + row['name'] + """\' .
             ?inc STO:has_summary ?summary .
           }
    """))
  if len(r) != 0:
    for index2, item in enumerate(r):
      #Add attibute
      summary_ind = basesto.Individual("event_individual_"+str(index)+"_"+str(index2))
      att = basesto.Sentence("att_event_"+str(index))
      summary_ind.hasAttribute.append(att)
      val = basesto.Value("val_event_"+str(index))
      att.hasAttributeValue.append(val)
      val.attributeValue.append(item[1])

      #Add attribution to individuals
      summary_inf = basesto.ElementaryInfon("event_"+str(index)+"_"+str(index2))
      sitname.supportedInfon.append(summary_inf)
      summary_rel = basesto.happenedInContext("happened_in_context_"+str(index)+"_"+str(index2))
      summary_inf.relation = summary_rel
      summary_inf.anchor1 = item[0]
      summary_inf.anchor2 = summary_ind

  #Pattern attack
  #attackPattern = basesto.INF("attack_pattern_"+str(index))
  #sitname.supportedInfon.append(attackPattern)

  r = list(default_world.sparql("""
           SELECT ?inc ?tech ?tact ?phase
           { ?inct rdf:type STO:Incidents_techniques .
             ?inct STO:uses_technique ?tech .
             ?inct STO:is_used_in ?inc .
             ?inc STO:has_name \'""" + row['name'] + """\' .
             ?tech STO:has_name ?techname .
             ?tech STO:is_in_tactic ?tact.
             ?tact STO:has_name ?tactname .
             ?tact STO:is_in_phase ?phase .
             ?phase STO:has_name ?phasename . }
    """))
  if len(r) > 1:
    for index2, item in enumerate(r):
      #AttackPattern phases
      attackPattern = basesto.ElementaryInfon("attack_pattern_phase_"+str(index)+"_"+str(index2))
      sitname.supportedInfon.append(attackPattern)
      attackPatternPhase = basesto.hasPhase("hasPhase_"+str(index)+"_"+str(index2))
      attackPattern.relation = attackPatternPhase
      attackPattern.anchor1 = item[0]
      attackPattern.anchor2 = item[3]
      #attackPattern.anchor2 = str(item[3])+"_"+str(item[2])+"_"+str(item[1])
      #cur_inc = Incidents(row['disarm_id'], namespace = onto, has_id = row['disarm_id'], has_name = row['name'], has_objecttype = row['objecttype'], has_summary = row['summary'], has_year_started = row['year_started'], has_attributions_seen = row['attributions_seen'], has_found_in_country = row['found_in_country'], has_urls = row['urls'], has_when_added = row['when_added'])

      #AttackPattern tactics
      attackPattern2 = basesto.ElementaryInfon("attack_pattern_tact_"+str(index)+"_"+str(index2))
      sitname.supportedInfon.append(attackPattern2)
      attackPatternTact = basesto.hasTactic("hasTactic_"+str(index)+"_"+str(index2))
      attackPattern2.relation = attackPatternTact
      attackPattern2.anchor1 = item[0]
      attackPattern2.anchor2 = item[2]

      #AttackPattern technics
      attackPattern3 = basesto.ElementaryInfon("attack_pattern_tech_"+str(index)+"_"+str(index2))
      sitname.supportedInfon.append(attackPattern3)
      attackPatternTech = basesto.hasTechnic("hasTechnic_"+str(index)+"_"+str(index2))
      attackPattern3.relation = attackPatternTech
      attackPattern3.anchor1 = item[0]
      attackPattern3.anchor2 = item[1]

  r = list(default_world.sparql("""
           SELECT ?inc ?tech ?tact ?phase ?countm
           { ?inc STO:has_name \'""" + row['name'] + """\' .
             ?inc STO:uses_technique ?tech .
             ?tech STO:has_name ?techname .
             ?tech STO:is_in_tactic ?tact.
             ?tact STO:has_name ?tactname .
             ?tact STO:is_in_phase ?phase .
             ?phase STO:has_name ?phasename .
             ?countm rdf:type STO:Countermeasures .
             ?countm STO:uses_tactic ?tact .
             ?countm STO:has_name ?countmname . }
    """))
  print(len(r))
  if len(r) > 1:
    print(len(r))
    print(r)
    for index2, item in enumerate(r):
      #CourseOfAction countermeasures
      CourseOfAction = basesto.ElementaryInfon("course_of_action_"+str(index)+"_"+str(index2))
      sitname.supportedInfon.append(CourseOfAction)
      CourseOfActionCnt = basesto.hasCountermeasure("hasCountermeasure_"+str(index)+"_"+str(index2))
      CourseOfAction.relation = CourseOfActionCnt
      CourseOfAction.anchor1 = item[0]
      CourseOfAction.anchor2 = item[4]

print(list(onto.individuals()))
```

# Mine Frequent Itemsets

Utils

```python
from collections import defaultdict
from itertools import chain, combinations

class Node:
    def __init__(self, itemName, frequency, parentNode):
        self.itemName = itemName
        self.count = frequency
        self.parent = parentNode
        self.children = {}
        self.next = None

    def increment(self, frequency):
        self.count += frequency

    def display(self, ind=1):
        print('  ' * ind, self.itemName, ' ', self.count)
        for child in list(self.children.values()):
            child.display(ind+1)

def constructTree(itemSetList, frequency, minSup):
    headerTable = defaultdict(int)
    # Counting frequency and create header table
    for idx, itemSet in enumerate(itemSetList):
        for item in itemSet:
            headerTable[item] += frequency[idx]

    # Deleting items below minSup
    headerTable = dict((item, sup) for item, sup in headerTable.items() if sup >= minSup)
    if(len(headerTable) == 0):
        return None, None

    # HeaderTable column [Item: [frequency, headNode]]
    for item in headerTable:
        headerTable[item] = [headerTable[item], None]

    # Init Null head node
    fpTree = Node('Null', 1, None)
    # Update FP tree for each cleaned and sorted itemSet
    for idx, itemSet in enumerate(itemSetList):
        itemSet = [item for item in itemSet if item in headerTable]
        itemSet.sort(key=lambda item: headerTable[item][0], reverse=True)
        # Traverse from root to leaf, update tree with given item
        currentNode = fpTree
        for item in itemSet:
            currentNode = updateTree(item, currentNode, headerTable, frequency[idx])

    return fpTree, headerTable

def updateHeaderTable(item, targetNode, headerTable):
    if(headerTable[item][1] == None):
        headerTable[item][1] = targetNode
    else:
        currentNode = headerTable[item][1]
        # Traverse to the last node then link it to the target
        while currentNode.next != None:
            currentNode = currentNode.next
        currentNode.next = targetNode

def updateTree(item, treeNode, headerTable, frequency):
    if item in treeNode.children:
        # If the item already exists, increment the count
        treeNode.children[item].increment(frequency)
    else:
        # Create a new branch
        newItemNode = Node(item, frequency, treeNode)
        treeNode.children[item] = newItemNode
        # Link the new branch to header table
        updateHeaderTable(item, newItemNode, headerTable)

    return treeNode.children[item]

def ascendFPtree(node, prefixPath):
    if node.parent != None:
        prefixPath.append(node.itemName)
        ascendFPtree(node.parent, prefixPath)

def findPrefixPath(basePat, headerTable):
    # First node in linked list
    treeNode = headerTable[basePat][1]
    condPats = []
    frequency = []
    while treeNode != None:
        prefixPath = []
        # From leaf node all the way to root
        ascendFPtree(treeNode, prefixPath)
        if len(prefixPath) > 1:
            # Storing the prefix path and it's corresponding count
            condPats.append(prefixPath[1:])
            frequency.append(treeNode.count)

        # Go to next node
        treeNode = treeNode.next
    return condPats, frequency

def mineTree(headerTable, minSup, preFix, freqItemList):
    # Sort the items with frequency and create a list
    sortedItemList = [item[0] for item in sorted(list(headerTable.items()), key=lambda p:p[1][0])]
    # Start with the lowest frequency
    for item in sortedItemList:
        # Pattern growth is achieved by the concatenation of suffix pattern with frequent patterns generated from conditional FP-tree
        newFreqSet = preFix.copy()
        newFreqSet.add(item)
        freqItemList.append(newFreqSet)
        # Find all prefix path, constrcut conditional pattern base
        conditionalPattBase, frequency = findPrefixPath(item, headerTable)
        # Construct conditonal FP Tree with conditional pattern base
        conditionalTree, newHeaderTable = constructTree(conditionalPattBase, frequency, minSup)
        if newHeaderTable != None:
            # Mining recursively on the tree
            mineTree(newHeaderTable, minSup,
                       newFreqSet, freqItemList)

def powerset(s):
    return chain.from_iterable(combinations(s, r) for r in range(1, len(s)))

def getSupport(testSet, itemSetList):
    count = 0
    for itemSet in itemSetList:
        if(set(testSet).issubset(itemSet)):
            count += 1
    return count

def associationRule(freqItemSet, itemSetList, minConf):
    rules = []
    for itemSet in freqItemSet:
        subsets = powerset(itemSet)
        itemSetSup = getSupport(itemSet, itemSetList)
        for s in subsets:
            confidence = float(itemSetSup / getSupport(s, itemSetList))
            if(confidence > minConf):
                rules.append([set(s), set(itemSet.difference(s)), confidence])
    return rules

def getFrequencyFromList(itemSetList):
    frequency = [1 for i in range(len(itemSetList))]
    return frequency
```

FPGrowth

```python
def fpgrowth(itemSetList, minSupRatio, minConf):
    frequency = getFrequencyFromList(itemSetList)
    minSup = len(itemSetList) * minSupRatio
    fpTree, headerTable = constructTree(itemSetList, frequency, minSup)
    if(fpTree == None):
        print('No frequent item set')
    else:
        freqItems = []
        mineTree(headerTable, minSup, set(), freqItems)
        rules = associationRule(freqItems, itemSetList, minConf)
        return freqItems, rules
```

```python
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import fpgrowth
import numpy as np
import warnings
import IPython
from IPython import get_ipython

# Ignora i DeprecationWarning
warnings.filterwarnings('ignore', category=DeprecationWarning)

# Configura IPython per gestire il codice asincrono correttamente
if IPython.version_info >= (7, 17):
    ip = get_ipython()
    if ip is not None:
        ip.config.TerminalInteractiveShell.autoawait = True
        ip.config.TerminalInteractiveShell.loop_runner = 'IPython.core.async_helpers._asyncio_runner'

# Carica il dataset
file_path = 'final_incidents_techniques_cleaned_3.csv'
df = pd.read_csv(file_path, sep=';')

# Create triple combinations
triples_list = df.apply(lambda row: f"{row['phase_id']}:{row['tactic_id']}:{row['technique_id']}", axis=1)
df['triple'] = triples_list
transactions = df.groupby('disarm_id_incident')['triple'].apply(list).tolist()

# Define internal and external utility values for each unique triple based on domain knowledge
internal_utilities = {}
external_utilities = {}

unique_triples = df['triple'].unique()

for triple in unique_triples:
    phase, tactic, technique = triple.split(':')

    # Example logic to assign utilities based on domain-specific knowledge
    internal_utilities[triple] = (
        0.9 if phase == 'P02' else
        0.7 if tactic == 'TA06' else
        0.5 if technique == 'T0018' else
        0.6
    )

    external_utilities[triple] = (
        0.8 if phase == 'P02' else
        0.6 if tactic == 'TA06' else
        0.4 if technique == 'T0018' else
        0.5
    )

# Define the membership function
def membership_function(quantity):
    low_threshold = 0.3
    middle_threshold = 0.6
    high_threshold = 1.0

    if quantity <= low_threshold:
        return {'Low': 1, 'Middle': 0, 'High': 0}
    elif quantity <= middle_threshold:
        low_value = (middle_threshold - quantity) / (middle_threshold - low_threshold)
        middle_value = (quantity - low_threshold) / (middle_threshold - low_threshold)
        return {'Low': low_value, 'Middle': middle_value, 'High': 0}
    else:
        middle_value = (high_threshold - quantity) / (high_threshold - middle_threshold)
        high_value = (quantity - middle_threshold) / (high_threshold - middle_threshold)
        return {'Low': 0, 'Middle': middle_value, 'High': high_value}

# Calculate fuzzy utility for an item
def calculate_fuzzy_utility(item):
    internal_utility = internal_utilities[item]
    external_utility = external_utilities[item]

    quantity = (internal_utility + external_utility) / 2
    p_x = 1  # Assuming unit profit for simplicity

    memberships = membership_function(quantity)
    fuzzy_utilities = {region: memberships[region] * quantity * p_x for region in memberships}
    total_fuzzy_utility = sum(fuzzy_utilities.values())

    return total_fuzzy_utility

# Transaction Encoder
te = TransactionEncoder()
te_ary = te.fit(transactions).transform(transactions)
df_trans = pd.DataFrame(te_ary, columns=te.columns_)

# FP-Growth Algorithm
frequent_itemsets = fpgrowth(df_trans, min_support=0.15, use_colnames=True)

# Calculate fuzzy utility and risk score for each itemset
def calculate_itemset_fuzzy_utility(itemset):
    total_fuzzy_utility = 0
    for item in itemset:
        total_fuzzy_utility += calculate_fuzzy_utility(item)
    return total_fuzzy_utility

frequent_itemsets['fuzzy_utility'] = frequent_itemsets['itemsets'].apply(lambda x: calculate_itemset_fuzzy_utility(list(x)))
frequent_itemsets['risk_score'] = frequent_itemsets['support'] * frequent_itemsets['fuzzy_utility']

# Sort by risk score
frequent_itemsets = frequent_itemsets.sort_values(by='risk_score', ascending=False)

# Display the top frequent itemsets with high risk scores
print(frequent_itemsets)

# Optional: Save the results to a CSV file
frequent_itemsets.to_csv('frequent_itemsets_with_risk_scores.csv', index=False)


# Output the first few frequent itemsets for inspection
frequent_itemsets.head(10)
```

```python
# Definizione delle utilit interne ed esterne (valori definiti da esperti di dominio)
internal_utilities = {
    'P03:TA08:T0046': 0.9,
    'P02:TA06:T0019': 0.8,
    'P02:TA15:T0010': 0.85,
    'P02:TA15:T0007': 0.7,
    'P03:TA10:T0057': 0.6,
    'P03:TA08:T0020': 0.75
}

external_utilities = {
    'P03:TA08:T0046': 0.8,
    'P02:TA06:T0019': 0.85,
    'P02:TA15:T0010': 0.9,
    'P02:TA15:T0007': 0.65,
    'P03:TA10:T0057': 0.55,
    'P03:TA08:T0020': 0.7
}

# Funzioni di membership per i termini linguistici (Low, Middle, High)
def membership_function(quantity):
    if quantity <= 1:
        return {'Low': 1, 'Middle': 0, 'High': 0}
    elif 1 < quantity <= 6:
        return {'Low': (6 - quantity) / 5, 'Middle': (quantity - 1) / 5, 'High': 0}
    elif 6 < quantity <= 11:
        return {'Low': 0, 'Middle': (11 - quantity) / 5, 'High': (quantity - 6) / 5}
    else:
        return {'Low': 0, 'Middle': 0, 'High': 1}

# Funzione per calcolare la fuzzy utility
def calculate_fuzzy_utility(itemset):
    fuzzy_util = 0
    for item in itemset:
        internal_utility = internal_utilities.get(item, 0)
        external_utility = external_utilities.get(item, 0)
        quantity = internal_utility * 10  # Esempio: utilizzo del valore dell'utilit interna come quantit per il calcolo della membership
        membership_values = membership_function(quantity)
        fuzzy_value = max(membership_values.values())  # Seleziona il valore massimo di membership
        fuzzy_util += fuzzy_value * internal_utility * external_utility
    return fuzzy_util / len(itemset)

# Calcolo dell'utilit fuzzy per ciascun itemset
frequent_itemsets['fuzzy_utility'] = frequent_itemsets.apply(
    lambda row: calculate_fuzzy_utility(row['itemsets']),
    axis=1
)

# Normalizzazione dei valori di fuzzy utility tra 0 e 1
min_fuzzy_utility = frequent_itemsets['fuzzy_utility'].min()
max_fuzzy_utility = frequent_itemsets['fuzzy_utility'].max()

frequent_itemsets['normalized_fuzzy_utility'] = (frequent_itemsets['fuzzy_utility'] - min_fuzzy_utility) / (max_fuzzy_utility - min_fuzzy_utility)

# Applicazione della soglia di taglio
threshold = 0.6
high_utility_itemsets = frequent_itemsets[frequent_itemsets['normalized_fuzzy_utility'] >= threshold]

# Ordinamento delle sequenze finali in base alla fuzzy utility normalizzata
high_utility_itemsets = high_utility_itemsets.sort_values(by='normalized_fuzzy_utility', ascending=False)

# Visualizzazione dei risultati finali
print(high_utility_itemsets)
```

```python
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import fpgrowth
import numpy as np
import warnings
import IPython
from IPython import get_ipython

# Ignora i DeprecationWarning
warnings.filterwarnings('ignore', category=DeprecationWarning)

# Configura IPython per gestire il codice asincrono correttamente
if IPython.version_info >= (7, 17):
    ip = get_ipython()
    if ip is not None:
        ip.config.TerminalInteractiveShell.autoawait = True
        ip.config.TerminalInteractiveShell.loop_runner = 'IPython.core.async_helpers._asyncio_runner'

# Carica il dataset
file_path =  'final_incidents_techniques_cleaned_3.csv'
df = pd.read_csv(file_path, sep=';')

# Create triple combinations
triples_list = df.apply(lambda row: f"{row['phase_id']}:{row['tactic_id']}:{row['technique_id']}", axis=1)
df['triple'] = triples_list
transactions = df.groupby('disarm_id_incident')['triple'].apply(list).tolist()

# Define internal and external utility values for each unique triple based on domain knowledge
# Define internal and external utility values for each unique triple
# These values are examples based on hypothetical DISARM framework assessments

# Definizione delle utility interna ed esterna per ogni item
internal_utilities = {
    'P02:TA05:T0018': 0.75,
    'P03:TA10:T0017': 0.70,
    'P02:TA06:T0019': 0.80,
    'P03:TA08:T0046': 0.85,
    'P02:TA15:T0007': 0.65,
    'P03:TA10:T0057': 0.60,
    'P02:TA07:T0029': 0.70,
    'P02:TA15:T0010': 0.75,
    'P02:TA07:T0043': 0.70,
    'P02:TA06:T0015': 0.65,
    'P03:TA08:T0039': 0.80,
    'P03:TA08:T0045': 0.75,
    'P03:TA08:T0020': 0.65,
    'P03:TA08:T0044': 0.70,
    'P02:TA05:T0016': 0.75,
    'P02:TA14:T0040': 0.80,
    'P03:TA17:T0049': 0.85,
    'P03:TA18:T0047': 0.70,
    'P03:TA18:T0048': 0.75,
    'P01:TA02:T0002': 0.70,
    'P02:TA16:T0011': 0.80,
    'P02:TA16:T0009': 0.75,
    'P02:TA06:T0023': 0.80,
    'P02:TA14:T0022': 0.75
}

external_utilities = {
    'P02:TA05:T0018': 0.70,
    'P03:TA10:T0017': 0.75,
    'P02:TA06:T0019': 0.85,
    'P03:TA08:T0046': 0.80,
    'P02:TA15:T0007': 0.65,
    'P03:TA10:T0057': 0.60,
    'P02:TA07:T0029': 0.75,
    'P02:TA15:T0010': 0.80,
    'P02:TA07:T0043': 0.70,
    'P02:TA06:T0015': 0.65,
    'P03:TA08:T0039': 0.85,
    'P03:TA08:T0045': 0.80,
    'P03:TA08:T0020': 0.65,
    'P03:TA08:T0044': 0.70,
    'P02:TA05:T0016': 0.75,
    'P02:TA14:T0040': 0.80,
    'P03:TA17:T0049': 0.85,
    'P03:TA18:T0047': 0.70,
    'P03:TA18:T0048': 0.75,
    'P01:TA02:T0002': 0.70,
    'P02:TA16:T0011': 0.80,
    'P02:TA16:T0009': 0.75,
    'P02:TA06:T0023': 0.80,
    'P02:TA14:T0022': 0.75
}

unique_triples = df['triple'].unique()

for triple in unique_triples:
    phase, tactic, technique = triple.split(':')

    # Example logic to assign utilities based on domain-specific knowledge
    internal_utilities[triple] = (
        0.9 if phase == 'P02' else
        0.7 if tactic == 'TA06' else
        0.5 if technique == 'T0018' else
        0.6
    )

    external_utilities[triple] = (
        0.8 if phase == 'P02' else
        0.6 if tactic == 'TA06' else
        0.4 if technique == 'T0018' else
        0.5
    )

# Define the membership function
def membership_function(quantity):
    low_threshold = 0.3
    middle_threshold = 0.6
    high_threshold = 1.0

    if quantity <= low_threshold:
        return {'Low': 1, 'Middle': 0, 'High': 0}
    elif quantity <= middle_threshold:
        low_value = (middle_threshold - quantity) / (middle_threshold - low_threshold)
        middle_value = (quantity - low_threshold) / (middle_threshold - low_threshold)
        return {'Low': low_value, 'Middle': middle_value, 'High': 0}
    else:
        middle_value = (high_threshold - quantity) / (high_threshold - middle_threshold)
        high_value = (quantity - middle_threshold) / (high_threshold - middle_threshold)
        return {'Low': 0, 'Middle': middle_value, 'High': high_value}

# Calculate fuzzy utility for an item
def calculate_fuzzy_utility(item):
    internal_utility = internal_utilities[item]
    external_utility = external_utilities[item]

    quantity = (internal_utility + external_utility) / 2
    p_x = 1  # Assuming unit profit for simplicity

    memberships = membership_function(quantity)
    fuzzy_utilities = {region: memberships[region] * quantity * p_x for region in memberships}
    total_fuzzy_utility = sum(fuzzy_utilities.values())

    return total_fuzzy_utility

# Transaction Encoder
te = TransactionEncoder()
te_ary = te.fit(transactions).transform(transactions)
df_trans = pd.DataFrame(te_ary, columns=te.columns_)

# FP-Growth Algorithm
frequent_itemsets = fpgrowth(df_trans, min_support=0.15, use_colnames=True)

# Calculate fuzzy utility and risk score for each itemset
def calculate_itemset_fuzzy_utility(itemset):
    total_fuzzy_utility = 0
    for item in itemset:
        total_fuzzy_utility += calculate_fuzzy_utility(item)
    return total_fuzzy_utility

frequent_itemsets['fuzzy_utility'] = frequent_itemsets['itemsets'].apply(lambda x: calculate_itemset_fuzzy_utility(list(x)))
frequent_itemsets['risk_score'] = round(frequent_itemsets['support'] * frequent_itemsets['fuzzy_utility'], 2)

# Sort by risk score
frequent_itemsets = frequent_itemsets.sort_values(by='risk_score', ascending=False)

# Display the top frequent itemsets with high risk scores
print(frequent_itemsets)

# Optional: Save the results to a CSV file
#frequent_itemsets.to_csv('frequent_itemsets_with_risk_scores.csv', index=False)

# Display the top frequent itemsets with high risk scores
#import ace_tools as tools
#tools.display_dataframe_to_user("Frequent Itemsets with High Risk Scores", frequent_itemsets)

# Output the first few frequent itemsets for inspection
frequent_itemsets.head(10)
```
